```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
import numpy as np
import pandas as pd
import bokeh.charts as bok
```

```python
df = pd.read_excel('ratio.xlsx')
df.columns.values[0] = 'ratio'
df.columns
colooms = df['ratio']
```

```python
df = df.iloc[:, 1:].transpose()
df.columns = colooms.values
df.head()
```

```python
print(df.columns.values.size)
print(df.dropna(how='all', axis='columns').columns.values.size)
```

```python
output_notebook()
plot = figure()

plot.circle(df['incomeperperson'], df['alcconsumption'], size=3)

show(plot)
```
